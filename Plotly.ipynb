```python
df = pd.DataFrame(np.random.randn(100,3), columns= ['A', 'B', 'C'])
df.head()
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 20
```

```python
df.shape
```

# Line Plot

```python
df.iplot()
```

```python
df.iplot(kind = 'area', fill = True)
```

```python
plt.plot(df)
```

```python
df.plot()
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 10)
```
